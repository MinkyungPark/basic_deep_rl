### HER + DDQN

```python
from collections import defaultdict
import matplotlib.pyplot as plt
import pickle

from train import bitflip_train

bits = 50  # more than 10^15 states
epochs = 40
stratgy='final'
success = defaultdict(list)

success['HER-DDQN-final'] = bitflip_train(bits, epochs, True, stratgy='final')
success['HER-DDQN-future'] = bitflip_train(bits, epochs, True, stratgy='future')
success['DDQN'] = bitflip_train(bits, epochs, False, stratgy=None)

for label, values in success.items():
    plt.plot(values, label=label)
    
plt.legend()
plt.xlabel('Epoch')
plt.ylabel('Success rate')
plt.title(f'Number of bits: {bits}')
plt.savefig(f'{bits}_bits.png', dpi=1000)
plt.show()
```
